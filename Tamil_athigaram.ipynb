```python
from textblob import TextBlob
import pandas as pd
d1=pd.DataFrame({"Tamil_meaning":l1[0]})
```

```python
d1
```

```python
d1["English_meaning"]=d1['Tamil_meaning'].apply(lambda x:TextBlob(x).translate(from_lang="ta",to="eng"))
```
