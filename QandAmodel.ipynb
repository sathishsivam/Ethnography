```python
import pandas as pd
from transformers import pipeline
```

```python
classifier=pipeline(task="table-question-answering")
```

```python
data=pd.read_html("https://en.wikipedia.org/wiki/National_Institutes_of_Technology_(India)",header=1)[2]
```

```python
data.columns
```

```python
data.drop(columns=["Website","No."],inplace=True)
```

```python
data
```

```python
data=data.astype(str)
```

```python
data
```

```python
query="What is the Name of the city of NIT Andhra"
```

```python
classifier(table=data,query=query)['answer']
```
